```python
# ============ CELL 1: Install ============


# ============ CELL 2: Imports ============
import torch
import torch.nn as nn
from transformers import (
    CLIPModel, CLIPProcessor,
    AutoModelForCausalLM, AutoTokenizer,
    BitsAndBytesConfig
)
from peft import LoraConfig, get_peft_model, TaskType
from datasets import load_dataset
from PIL import Image
import gradio as gr

# ============ CELL 3: Load CLIP ============
checkpoint = "openai/clip-vit-large-patch14"
clip_model = CLIPModel.from_pretrained(checkpoint)
clip_processor = CLIPProcessor.from_pretrained(checkpoint)
clip_model.eval()
for param in clip_model.parameters():
    param.requires_grad = False


```

```python
# ============ CELL 4: Image feature function ============
def get_image_features(image):
    image = image.convert("RGB")
    inputs = clip_processor(images=image, return_tensors="pt")
    with torch.no_grad():
        features = clip_model.get_image_features(**inputs)
    return features

# ============ CELL 5: Projection layer ============
class ProjectionLayer(nn.Module):
    def __init__(self):
        super().__init__()
        self.proj = nn.Linear(768, 2048)

    def forward(self, x):
        return self.proj(x)

projection = ProjectionLayer()
```

```python
# ============ CELL 6: Dataset ============
dataset = load_dataset("hf-vision/chest-xray-pneumonia", split="train", streaming=True)

def label_to_qa(label):
    if label == 0:
        return {"question": "Is pneumonia present in this X-ray?",
                "answer": "No, pneumonia is not detected."}
    else:
        return {"question": "Is pneumonia present in this X-ray?",
                "answer": "Yes, pneumonia is detected."}

# ============ CELL 7: Load LLM with 4-bit ============
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

llm_checkpoint = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
llm_model = AutoModelForCausalLM.from_pretrained(
    llm_checkpoint,
    quantization_config=bnb_config,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(llm_checkpoint)
```

```python
# ============ CELL 8: Apply LoRA ============
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type=TaskType.CAUSAL_LM
)
llm_model = get_peft_model(llm_model, lora_config)
llm_model.print_trainable_parameters()

# ============ CELL 9: Optimizer ============
device = next(llm_model.parameters()).device
projection = projection.to(device)

optimizer = torch.optim.AdamW(
    list(filter(lambda p: p.requires_grad, llm_model.parameters())) +
    list(projection.parameters()),
    lr=2e-4
)
```

```python
# ============ CELL 10: Training loop ============
for step, example in enumerate(dataset):
    image = example["image"]
    label = example["label"]
    qa = label_to_qa(label)

    features = get_image_features(image).to(device)
    image_embedding = projection(features)
    image_embedding = image_embedding.unsqueeze(1)

    text = qa["question"] + " " + qa["answer"]
    inputs = tokenizer(text, return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}

    text_embeddings = llm_model.base_model.model.model.embed_tokens(inputs["input_ids"])

    image_embedding = image_embedding.to(device)
    combined = torch.cat([image_embedding, text_embeddings], dim=1)
    combined = combined.to(dtype=llm_model.dtype)

    image_label = torch.tensor([[-100]]).to(device)
    labels = torch.cat([image_label, inputs["input_ids"]], dim=1)

    outputs = llm_model(inputs_embeds=combined, labels=labels)
    loss = outputs.loss

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if step % 10 == 0:
        print(f"Step {step} Loss: {loss.item()}")
    if step == 100:
        break
```

```python
# ============ CELL 11: Save ============
llm_model.save_pretrained("./xray_vlm")
torch.save(projection.state_dict(), "./xray_vlm/projection.pt")

# ============ CELL 12: Inference function ============
def answer_question(image, question):
    features = get_image_features(image).to(device)
    image_embedding = projection(features)
    image_embedding = image_embedding.unsqueeze(1)
    image_embedding = image_embedding.to(device)

    inputs = tokenizer(question, return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}
    text_embeddings = llm_model.base_model.model.model.embed_tokens(inputs["input_ids"])

    combined = torch.cat([image_embedding, text_embeddings], dim=1)
    combined = combined.to(dtype=llm_model.dtype)

    attention_mask = torch.ones(combined.shape[:2], dtype=torch.long, device=device)

    with torch.no_grad():
        outputs = llm_model.generate(
            inputs_embeds=combined,
            attention_mask=attention_mask,
            max_new_tokens=20,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return answer
```

```python

# ============ CELL 13: Gradio UI ============
demo = gr.Interface(
    fn=answer_question,
    inputs=[
        gr.Image(type="pil"),
        gr.Textbox(placeholder="Ask a question about the X-ray...")
    ],
    outputs=gr.Textbox(label="Model Answer"),
    title="X-Ray VLM",
    description="Upload a chest X-ray and ask a question!"
)

demo.launch(share=True, debug=True)
```
